```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")

#Data_Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Rather than deleting the rows where average_year_salary and average_rate are NaN,
#let's fill them with median values

#First, let's calculate the median value for average salary year
median_salary_year = df['salary_year_avg'].median()
```

```python
#salary_hour_avg median value
median_salary_hour = df['salary_hour_avg'].median()
```

```python
#before modifying the dataset, let's duplicate it

df_filled = df.copy() 
```

```python
#Setting new column in duplicated dataset
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
```

```python
#same thing for salary_hour column
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled
```

```python
#Great done, perfect
#Let's continue
#I will remove duplicate but first, I will create a duplicated of df_filled

df_unique = df_filled.copy()

df_unique = df_unique.drop_duplicates()

# This pretty cool code displays how many rows were dropped

print('Lenght of original dataset :', len(df_filled))
print('Lenght of unique dataset   :', len(df_unique))
print('Number of dropped rows     :', len(df_filled) - len(df_unique))
```

```python
#We can run duplicates on specific columns
df_unique = df_unique.drop_duplicates(subset =['job_title', 'company_name'])

print('Lenght of original dataset :', len(df_filled))
print('Lenght of unique dataset   :', len(df_unique))
print('Number of dropped rows     :', len(df_filled) - len(df_unique))
```
