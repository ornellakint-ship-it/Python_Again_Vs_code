```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")

#Data_Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#removing the null values from the dataset in  the salary_year_avg column

df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
```

```python
#Now we are going to make a calculation on all the values in salary_avg_year
#We created a function

df_salary = df[pd.notna(df['salary_year_avg'])].copy()

def projected_salary(salary):
    return salary * 1.03


df_salary['salary_year_avg'].apply(projected_salary)
```

```python
#Displaying calculated and normal salaries sides by sides

df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)

df_salary[['salary_year_avg', 'salary_year_inflated']]
```
